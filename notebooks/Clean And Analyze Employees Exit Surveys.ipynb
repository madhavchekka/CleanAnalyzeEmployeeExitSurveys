Now, before we start cleaning and manipulating the rest of our data, let's verify that the data doesn't contain any major inconsistencies (to the best of our knowledge).Because even a good analysis on bad data is a bad analysis. Lets start with cease_date, dete_start_date, and role_start_date fields in dete_survey data. 


Since the cease_date is the last year of the person's employment and the dete_start_date is the person's first year of employment, it wouldn't make sense to have years after the current date.

Given that most people in this field start working in their 20s, it's also unlikely that the dete_start_date was before the year 1940.

Also, role_start_date cannot be before the year of dete_start_date. cease_date cannot be before role_start_date or dete_start_date. Therefore in the next steps, dropped two records of such. Looking at the boxplots, dete_start_date values span from 1963 till 2013 and cease_date values span from 2006 till 2014.

```python
combined_updated['institute_service'].astype(str).str.extract(pattern)['year_from'].astype(float)
```

```python
def set_service_cat(value):
    if pd.isnull(value):
        return np.nan
    elif 0 <= value < 3:
        return 'New'
    elif 3<= value <= 6:
        return 'Experienced'
    elif 7 <= value <= 10:
        return 'Established'
    elif value >= 11:
        return 'Veteran'
    else:
        pass
```

```python
combined_updated.loc[:,'service_cat'] = combined_updated['institute_service'].astype(str).str.extract(pattern)['year_from'].astype(float).apply(set_service_cat).copy()
```
